```
test["comp1_inv"].fillna(0.0, inplace = True)
test["comp1_rate_percent_diff"].fillna(0.0, inplace = True)
test["comp2_rate"].fillna(0.0, inplace = True)
test["comp2_inv"].fillna(0.0, inplace = True)
test["comp2_rate_percent_diff"].fillna(0.0, inplace = True)
test["comp3_rate"].fillna(0.0, inplace = True)
test["comp3_inv"].fillna(0.0, inplace = True)
test["comp3_rate_percent_diff"].fillna(0.0, inplace = True)
test["comp4_rate"].fillna(0.0, inplace = True)
test["comp4_inv"].fillna(0.0, inplace = True)
test["comp4_rate_percent_diff"].fillna(0.0, inplace = True)
test["comp5_rate"].fillna(0.0, inplace = True)
test["comp5_inv"].fillna(0.0, inplace = True)
test["comp5_rate_percent_diff"].fillna(0.0, inplace = True)
test["comp6_rate"].fillna(0.0, inplace = True)
test["comp6_inv"].fillna(0.0, inplace = True)
test["comp6_rate_percent_diff"].fillna(0.0, inplace = True)
test["comp7_rate"].fillna(0.0, inplace = True)
test["comp7_inv"].fillna(0.0, inplace = True)
test["comp7_rate_percent_diff"].fillna(0.0, inplace = True)
test["comp8_rate"].fillna(0.0, inplace = True)
test["comp8_inv"].fillna(0.0, inplace = True)
test["comp8_rate_percent_diff"].fillna(0.0, inplace = True)
test.head(5)
```

```
test["orig_destination_distance"].fillna(test["orig_destination_distance"].median(), inplace = True)
```

```
test["srch_query_affinity_score"].fillna(0.0, inplace = True)
```

```
test["prop_review_score"].fillna(test["prop_review_score"].median(), inplace = True)
```

```
test["prop_location_score2"].fillna(test["prop_location_score2"].min(), inplace = True)
```

```
test["visitor_hist_starrating"].fillna(test["visitor_hist_starrating"].median(), inplace = True)
```

```
test["visitor_hist_adr_usd"].fillna(test["visitor_hist_adr_usd"].median(), inplace = True)
```

```
test["popularity_score"].fillna(test["popularity_score"].median(), inplace=True)
```

```
test.isnull().sum()
```

```
test = test.drop_duplicates()
```

```
test['histpricediff'] = (test["visitor_hist_adr_usd"] - test["price_usd"])
```

```
test["histstardiff"] = (test["visitor_hist_starrating"] - test["prop_starrating"])
```

```
means = test.groupby(test.srch_id)["price_usd"].transform('mean')
test["pricediff"] = test["price_usd"] - means
means.drop(means.index, inplace=True)
```

```
means = test.groupby(test.srch_id)["prop_starrating"].transform('mean')
test["stardiff"] = test["prop_starrating"] - means
means.drop(means.index, inplace=True)
```

```
means = test.groupby(test.srch_id)["prop_review_score"].transform('mean')
test["reviewscorediff"] = test["prop_review_score"] - means
means.drop(means.index, inplace=True)
```

```
means = test.groupby(test.srch_id)["prop_location_score1"].transform('mean')
test["locationscore1diff"] = test["prop_location_score1"] - means
means.drop(means.index, inplace=True)
```

```
means = test.groupby(test.srch_id)["prop_location_score2"].transform('mean')
test["locationscore2diff"] = test["prop_location_score2"] - means
means.drop(means.index, inplace=True)
```

```
test = test.sort_values(by=["prop_id","date_time","srch_id"])
```

```
test['pricechange'] = test.groupby(['prop_id']).price_usd.pct_change()
```

```
test["pricechange"].fillna(0.0, inplace = True)
```

```
test = test.sort_values(by=["srch_id","date_time"])
```

```
tz = test.groupby("prop_id")["srch_id"].agg("count").reset_index()
tz.head(5)
```

```
test["hotel_count"] = test["prop_id"].map(tz.set_index("prop_id")["srch_id"])
test.head(5)
```

```
test['is219'] = [1 if value==219 else 0 for value in test['visitor_location_country_id']]
```

```
test["usd_diff"] = (test["visitor_hist_adr_usd"] - test["price_usd"]).abs() 
```

```
test["prop_starrating"].mean()
```

```
test["prop_starrating_monotonic"] = (test["prop_starrating"] - 3.1825417614151363).abs()
```

```
test["room_count_booking_window"] = test["srch_booking_window"]*test["srch_room_count"]
```

```
test["people_stay_count"] = test["srch_adults_count"] + test["srch_children_count"]
```

```
test['month'] = pd.DatetimeIndex(test['date_time']).month
```

```
test.head(5)
```

```
test["price_usd"] = np.log10(test["price_usd"])
```

```
test.drop(["date_time","srch_adults_count","srch_children_count","srch_room_count","srch_booking_window","prop_starrating"], axis=1, inplace=True)
```
