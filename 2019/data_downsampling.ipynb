```
import pandas as pd
import numpy as np
df = pd.read_csv("/content/drive/My Drive/Colab Notebooks/data_preprocessing2.csv", sep = ',')
df.head(5)
```

```
df.drop(["target"], axis=1, inplace=True)
```

```
target_column = "target"
conditions = [df["click_bool"] == 0, (df["click_bool"] == 1) & (df["booking_bool"] == 0), df["booking_bool"] == 1]
choices = ["None", "Click", "Booked"]
df[target_column] = np.select(conditions, choices, default="None")
```

```
df[df.click_bool == 1].head(5)
```

```
df[df.target == "None"].count()
```

```
df_0 = df[df.click_bool == 0]
df_1 = df[df.click_bool == 1]
```

```
df_0["srch_id"].count()
```

```
df_1["srch_id"].count()
```

```
df_0 = df_0.sample(221879)
```

```
df_0["srch_id"].count()
```

```
df = pd.concat([df_0,df_1])
```

```
df["srch_id"].count()
```

```
df = df.sort_values(by=["srch_id","month","click_bool","booking_bool"])
df.head(5)
```
